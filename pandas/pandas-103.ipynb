```python
import pandas as pd
df = pd.read_csv('pokemon.csv')
```

## Conditional Change

```python
# get row from condition, select col (Type 1) and assign value
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer (fire)'
df
```

```python
df.loc[df['Type 1'] == 'Flamer (fire)', 'Type 1'] = 'Fire'
df
```

```python
# set fire type poke to legendary 
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True
df.loc[df['Type 1'] == 'Fire'].head()
```
