```python
import pandas as pd
df = pd.read_csv('pokemon.csv')
```

```python
# sort/describe
df.describe()
```

```python
df.sort_values(['Type 1', 'HP'], ascending=[True, False])
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head()
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head()
```

```python
df_new = df[df['Type 1'] == 'Fire']
df_new.head()
```

```python
df_new.to_csv('Fire.csv', index=False)
```

```python
df_new.to_excel('Fire.xlsx')
```
